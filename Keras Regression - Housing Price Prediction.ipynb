```python
x=df.drop('price',axis=1).values
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
xtrain
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
```

```python
len(xtrain)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model=Sequential()
```

```python
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse') #Regression
```

```python

model.fit(x=xtrain,y=ytrain.values,
          validation_data=(xtest,ytest.values),
          batch_size=80,epochs=450)
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
loss.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,explained_variance_score
```

```python
pred=model.predict(xtest)
```

```python
pred
```

```python
mean_absolute_error(ytest.values,pred)
```

```python
mean_squared_error(ytest.values,pred)
```
